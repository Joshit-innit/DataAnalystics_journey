```python
x = dataset.iloc[:,[3,4]].values
x = pd.DataFrame(x)
x.head()
```

now how to choose the number of cluster? 
We use the WCSS (Within cluster sum of squares) 

To choose the best number of clusters in K-Means clustering, we use the WCSS (Within-Cluster Sum of Squares) method, also known as the Elbow Method.

WCSS measures how close the data points are to their cluster centroid. Lower WCSS means better clustering because the points are tightly grouped around the center.

1. Run K-Means for different values of k
2. Calculate WCSS for each value
3. Plot a graph of:
    X-axis → Number of clusters
    Y-axis → WCSS

The graph usually forms an “arm-like” shape. The point where the curve starts slowing down is called the Elbow Point. That value is considered the optimal number of clusters because adding more clusters after that gives very little improvement.


```python
wcss = []
for i in range(1,11):
    kmeans = KMeans(n_clusters=i, init='k-means++', random_state=123)
    kmeans.fit(x)
    wcss.append(kmeans.inertia_)

    
```

```python
sns.set()
plt.plot(range(1,11), wcss)
plt.title("The elbow graph")
plt.xlabel("Number of clusters")
plt.ylabel("WCSS")
plt.show()
```

```python
kmeans = KMeans(n_clusters=5, init='k-means++', random_state=123)

#returning a label for each dataset point based upon the cluster
y = kmeans.fit_predict(x)
y = pd.DataFrame(y)
y.head()
```

```python
# Convert y to 1D array
y = y.ravel()

plt.figure(figsize=(10,10))

plt.scatter(x[y==0,0], x[y==0,1], s=50, c='green', label='Cluster 1')
plt.scatter(x[y==1,0], x[y==1,1], s=50, c='red', label='Cluster 2')
plt.scatter(x[y==2,0], x[y==2,1], s=50, c='blue', label='Cluster 3')
plt.scatter(x[y==3,0], x[y==3,1], s=50, c='yellow', label='Cluster 4')
plt.scatter(x[y==4,0], x[y==4,1], s=50, c='pink', label='Cluster 5')

# Plot centroids
plt.scatter(
    kmeans.cluster_centers_[:,0],
    kmeans.cluster_centers_[:,1],
    s=200,
    c='black',
    marker='X',
    label='Centroids'
)

plt.title("Customer Groups")
plt.xlabel("Annual Income")
plt.ylabel("Spending Score")

plt.legend()
plt.show()
```
