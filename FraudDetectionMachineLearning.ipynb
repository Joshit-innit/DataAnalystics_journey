As the dataset is highly imbalanced, we use a technique called "undersampling".

> Undersampling : Undersampling reduces the number of samples in the majority class to balance the dataset.

Undersampling helps the model:

1. Pay attention to minority classes
2. Improve recall for rare events
3. Reduce bias toward majority class

It is commonly used in:

1. Fraud detection
2. Disease prediction
3. Spam detection
4. Anomaly detection

here in this dataset : 
Class
0    284315 -> legit
1       492 -> fraud


```python
legit_sample = legit.sample(n=492) # takes 492 legit valus
new_dataset = pd.concat([legit_sample, fraud], axis = 0) # concatination of fruad and legit values
new_dataset.head()
```

```python
new_dataset.groupby('Class').mean()
```

```python
x = new_dataset.drop(columns=['Class'], axis = 1)
y = new_dataset['Class']
print(x)
print(y)
```

```python
x_test, x_train, y_test, y_train = train_test_split(x, y, random_state = 23,test_size=0.2, stratify=y)
```

```python
model = LogisticRegression()
model.fit(x_train, y_train)
```

```python
x_train_prediction = model.predict(x_train)
training_prediction = accuracy_score(x_train_prediction, y_train)
print("The training prediction score is : ", training_prediction * 100)

x_test_prediction = model.predict(x_test)
testing_prediction = accuracy_score(x_test_prediction, y_test)
print("The testing prediction score is :", testing_prediction * 100)
```
