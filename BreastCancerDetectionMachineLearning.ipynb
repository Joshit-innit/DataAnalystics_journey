```python
# Importing NumPy library
# Used for numerical computations and array operations

import numpy as np


# Importing pandas library
# Used for data loading, manipulation, and analysis

import pandas as pd


# Importing seaborn library
# Used for statistical data visualization

import seaborn as sns


# Importing matplotlib.pyplot
# Used for plotting graphs and charts

import matplotlib.pyplot as plt


# Importing LabelEncoder
# Used to convert categorical text labels into numerical values

from sklearn.preprocessing import LabelEncoder


# Importing train_test_split
# Used to split dataset into training data and testing data

# Importing GridSearchCV
# Used for hyperparameter tuning and finding best model parameters

from sklearn.model_selection import train_test_split, GridSearchCV


# Importing LogisticRegression model
# Used for binary and multiclass classification problems

from sklearn.linear_model import LogisticRegression


# Importing accuracy_score
# Used to evaluate model accuracy

from sklearn.metrics import accuracy_score, classification_report


# Confirmation message after successful imports

print("imported")
```

```python
# Loading the breast cancer dataset from the CSV file using pandas

dataset = pd.read_csv(
    "/kaggle/input/datasets/yasserh/breast-cancer-dataset/breast-cancer.csv"
)

# Displaying the first 5 rows of the dataset
# Helps understand:
# - column names
# - dataset structure
# - sample records
# - feature values

dataset.head()
```

```python
# Checking the number of missing/null values in each column of the dataset

# isnull() -> identifies missing values
# sum() -> counts total missing values column-wise

dataset.isnull().sum()
```

```python
# Displaying the shape of the dataset

# shape returns:
# (number_of_rows, number_of_columns)

dataset.shape
```

```python
# Generating statistical summary of numerical columns in the dataset

# describe() provides important statistical information such as:
# - count  -> total non-null values
# - mean   -> average value
# - std    -> standard deviation
# - min    -> minimum value
# - 25%    -> first quartile
# - 50%    -> median value
# - 75%    -> third quartile
# - max    -> maximum value

dataset.describe()
```

```python
# Calculating correlation between numerical features

# Dropping the 'id' column because:
# - it is only a unique identifier
# - it does not contribute meaningful information for analysis

correlation = dataset.drop(
    columns=['id'],
    axis=1
).corr(numeric_only=True)

# Setting figure size for better visualization

plt.figure(figsize=(12,10))

# Creating heatmap for correlation matrix

sns.heatmap(
    
    correlation,
    
    cmap='coolwarm'
    # Defines color theme for correlation values
)

# Adding title to the heatmap

plt.title("Correlation of the variables")

# Displaying the heatmap

plt.show()
```

```python
# Grouping the dataset based on diagnosis values
# and calculating the average of all numerical features

# diagnosis categories:
# B -> Benign
# M -> Malignant

# groupby('diagnosis')
# Separates data based on diagnosis category

# mean()
# Calculates average value of each numerical feature
# for both diagnosis groups

dataset.groupby('diagnosis').mean(numeric_only=True)
```

```python
# Creating an object of LabelEncoder
# Used to convert categorical labels into numerical values

encoder = LabelEncoder()

# Encoding the 'diagnosis' column
#
# Converts:
# M -> 1  (Malignant -> cancer cells present)
# B -> 0  (Benign -> non-cancerous)
#
# Machine learning models require numerical labels,
# so encoding is necessary

dataset['diagnosis'] = encoder.fit_transform(dataset['diagnosis'])

# Displaying first 5 rows after encoding

dataset.head()


# Label Mapping:
# M -> 1  -> cancer cells present
# B -> 0  -> no cancer cells
```

```python
# Splitting the dataset into features (input data) and target (output label)

# x contains input features used for prediction

x = dataset.drop(
    columns=['id', 'diagnosis']
)

# 'id' column is removed because:
# - it is only a unique identifier
# - it does not help in prediction

# 'diagnosis' column is removed because:
# - it is the target/output variable
# - the model should predict this column


# y contains target labels
#
# 0 -> Benign (non-cancerous)
# 1 -> Malignant (cancerous)

y = dataset['diagnosis']
```

```python
# Splitting the dataset into training data and testing data

# x -> input features
# y -> target labels

# test_size=0.2
# 20% of data is used for testing
# 80% of data is used for training

# random_state=123
# Ensures reproducible/randomly consistent splitting

x_train, x_test, y_train, y_test = train_test_split(
    
    x,
    y,
    
    random_state=123,
    
    test_size=0.2
)

# Displaying shapes of training and testing datasets

# x_train.shape -> training feature dimensions
# x_test.shape  -> testing feature dimensions

print(x_train.shape, x_test.shape)

# y_train.shape -> training label dimensions
# y_test.shape  -> testing label dimensions

print(y_train.shape, y_test.shape)
```

```python
# Creating an object of Logistic Regression model

# Logistic Regression is used for classification problems
# In this project:
# 0 -> Benign (non-cancerous)
# 1 -> Malignant (cancerous)

# max_iter=1000
# Increases the maximum number of training iterations
# Helps the solver converge properly and avoids ConvergenceWarning

model = LogisticRegression(max_iter=1000)
```

```python
# Creating different hyperparameter combinations for Logistic Regression

param_grid = {
    
    # C controls regularization strength
    # Smaller values -> stronger regularization
    # Larger values  -> weaker regularization
    
    'C': [0.01, 0.1, 1, 10],

    
    # solver specifies the optimization algorithm
    # liblinear -> good for small datasets
    # lbfgs     -> general-purpose optimizer
    
    'solver': ['liblinear', 'lbfgs']
}


# Creating GridSearchCV object

grid_search = GridSearchCV(
    
    # estimator -> machine learning model to optimize
    
    estimator=model,

    
    # param_grid -> hyperparameter combinations to test
    
    param_grid=param_grid,

    
    # scoring='accuracy'
    # Evaluates model performance using accuracy score
    
    scoring='accuracy',

    
    # cv=5
    # Performs 5-fold cross-validation
    
    cv=5,

    
    # n_jobs=-1
    # Uses all CPU cores for faster processing
    
    n_jobs=-1
)


# Training GridSearchCV on training data
#
# GridSearchCV will:
# 1. Try different parameter combinations
# 2. Train multiple models
# 3. Evaluate performance using cross-validation
# 4. Select the best hyperparameters

grid_search.fit(x_train, y_train)


# Displaying the best hyperparameters found by GridSearchCV

print("Best parameter : ", grid_search.best_params_)
```

```python
# Retrieving the best trained model from GridSearchCV

best_model = grid_search.best_estimator_


# ---------------- TRAINING ACCURACY ---------------- #

# Making predictions on training data

train_prediction = best_model.predict(x_train)

# Calculating training accuracy
#
# accuracy_score(actual_values, predicted_values)

training_accuracy = accuracy_score(
    
    y_train,
    
    train_prediction
)

# Displaying training accuracy percentage

print(training_accuracy * 100)

print(classification_report(y_train, train_prediction))


# ---------------- TESTING ACCURACY ---------------- #

# Making predictions on testing data

test_prediction = best_model.predict(x_test)

# Calculating testing accuracy

testing_accuracy = accuracy_score(
    
    y_test,
    
    test_prediction
)

# Displaying testing accuracy percentage

print(testing_accuracy * 100)
print(classification_report(y_test, test_prediction))
```

```python
# ---------------- PREDICTIVE SYSTEM ---------------- #

# Taking one sample input data for prediction
# The values represent feature measurements of a tumor

input_data = (
     12.32, 12.39, 78.85, 464.1, 0.1028,
    0.06981, 0.03987, 0.03700, 0.1959, 0.05955,
    0.2360, 0.6656, 1.670, 17.43, 0.008045,
    0.01180, 0.01683, 0.01241, 0.01924, 0.002248,
    13.50, 15.64, 86.97, 549.1, 0.1385,
    0.1266, 0.1242, 0.09391, 0.2827, 0.06771
)

# Converting input data into NumPy array

input_data_as_numpy_array = np.asarray(input_data)

# Reshaping the array because the model expects input in 2D format
#
# reshape(1, -1):
# 1  -> one data sample/row
# -1 -> automatically determines number of columns/features

input_data_reshaped = input_data_as_numpy_array.reshape(1, -1)

# Making prediction using the trained best model

prediction = best_model.predict(input_data_reshaped)

# Interpreting prediction result

if prediction[0] == 1:
    
    print("The tumor is Malignant (Cancerous)")
    
else:
    
    print("The tumor is Benign (Non-Cancerous)")
    

prediction_probability = best_model.predict_proba(input_data_reshaped)

print(prediction_probability)
```
